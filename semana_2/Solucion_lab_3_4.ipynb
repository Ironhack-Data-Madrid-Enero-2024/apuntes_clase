But this time, we got new data, which can be found in the following 2 CSV files located at the following links: 
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv

Combine the data from the three dataframes into a single dataframe, named "customer_data", using appropriate merging, concatenating, and joining techniques.

Verify that the customer_data dataframe contains all the rows and columns from the three original dataframes.

```python
import pandas as pd

# ense√±e todas las columnas y filas
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1_clean.csv')

df1.head()
```

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv')


df2.columns = [e.lower().replace(' ', '_') for e in df2.columns] 


df2.rename(columns={'st': 'state'}, inplace=True)


df2.head()
```

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv')

df3.columns = [e.lower().replace(' ', '_') for e in df3.columns] 


df3.rename(columns={'st': 'state'}, inplace=True)


df3.head()
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)

df.head()
```

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')

df.head()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
df.pivot_table(values = 'total_claim_amount',
               index = 'sales_channel',
               aggfunc='sum').round(2).reset_index()
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
df.pivot_table(values = 'customer_lifetime_value',
               index = ['gender', 'education'],
               aggfunc='mean').round(2).reset_index()
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign. How many customers were there?

2. Using the original dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
df.head()
```

```python
yes = df[df.response=='Yes']

yes.head()
```

```python
df[df.response=='Yes'].groupby(['policy_type', 'gender'])['total_claim_amount'].mean().reset_index()
```

6. Display a new DataFrame that contains the number of policies sold by month by state for the top 3 states with the highest number of policies sold.

*Hint:* 
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
grupo = df.groupby(['state', 'month'])['number_of_policies'].count().reset_index()
```
